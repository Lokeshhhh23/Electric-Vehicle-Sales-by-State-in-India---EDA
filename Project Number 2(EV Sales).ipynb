**Exploratory Data Analysis(EDA) of  Electric Vehicle Sales by State in India**

 This project aims to analyze and predict the sales of Electric Vehicles (EV) by state in
 India using machine learning. The dataset contains the following columns :

1. Year: The year of the sales.
2. Month_Name: The month in which sales occurred.
3. Date: The specific date of the sales.
4. State: The state in India where the sales occurred.
5. Vehicle_Class: The class of the vehicle (e.g.,sedan,SUV,etc.).
6. Vehicle_Category: The category of the vehicle (e.g.,commercial,passenger).
7. Vehicle_Type: The type of the vehicle (e.g.,2-wheeler,4-wheeler).
8. EV_Sales_Quantity: The quantity of EVsales.

```python
df['Year'] = df['Year'].astype(int)
```

```python
categorical_columns = ['Month_Name','State','Vehicle_Class','Vehicle_Category','Vehicle_Type']
df[categorical_columns]= df[categorical_columns].astype('category')
```

```python
df.info()
```

**Data Visualisation**

```python
plt.figure(figsize=(6,4))
plt.title('Yearly Analysis of EY Sales in India')
sns.lineplot(x='Year',y='EV_Sales_Quantity',data = df  ,marker = 'o', color = 'b')
plt.xlabel('Year')
plt.ylabel('EV Sales')
plt.show()
```

As seen in this the EV Sales got increase after 2020 due
to a combination of factors, including 
government incentives, environmental concerns, technological advancements, and growing public interest
Electric two-wheelers led the charge, dominating the market with nearly 60% of 
the total electric vehicle sales in India .

```python
plt.figure(figsize=(6,4))
plt.title('Monthly Analysis of EV Sales in India')
sns.lineplot(x = 'Month_Name',y = 'EV_Sales_Quantity', data = df , marker = 'o',color = 'r')
plt.xlabel('Month')
plt.ylabel('EV Sales')
plt.show()
```

```python
plt.figure(figsize=(6,7))
plt.title('State-Wise Analysis of EV Sales')
sns.barplot(y=df['State'],x=df['EV_Sales_Quantity'], data = df ,hue ='State' , palette = 'bright')
plt.xlabel('States')
plt.ylabel('EV Sales');
```

After seen this we get to know about the Uttar Pradesh is the State where maximum Number of EV Sales 
happens.

```python
plt.figure(figsize=(15,4))
plt.title('Analysis by Vehicle Class')
sns.barplot(x='Vehicle_Class',y='EV_Sales_Quantity',data = df,hue ='Vehicle_Class' , palette = 'bright')
plt.xlabel('Vehicle Class')
plt.ylabel('EV Sales')
plt.xticks(rotation = 90);
```

By this Analysis we get to know about the major 2 Categorys of vehicle class that have maximum number 
of EV Sales are first is E-Rickshaw and M-Cycle/Scooter . 
By this we have insight is maximum number of E-Rickshaw sold in the Uttar Pradesh . 

```python
plt.figure(figsize=(6,7))
plt.title('Analysis of Vehicle Category')
sns.barplot(y='EV_Sales_Quantity',x='Vehicle_Category',data = df,hue ='Vehicle_Category' , palette = 'bright')
plt.xlabel('Vehicle Category')
plt.ylabel('EV Sales')
plt.xticks(rotation=0);
```

The majority of vehicles which gives us sales of 2 and 3 - Wheelers .
There is small difference between them . 

```python
plt.figure(figsize=(6,4))
plt.title('Analysis Vehicle Type')
sns.barplot(y='EV_Sales_Quantity',x='Vehicle_Type',data = df,hue ='Vehicle_Type' , palette = 'bright')
plt.xlabel('Vehicle Type')
plt.ylabel('EV Sales')
plt.xticks(rotation=90);
```

Through this we get to know the vehicle type which is majorily used is 3-Wheeler shared (Low Speed)Vehicle 
after that the 2 - Wheeler is used .
